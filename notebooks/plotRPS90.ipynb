```python
import os
import sys
topFolder = os.getcwd().split('\\').pop(-1)
moduleDirectory = os.getcwd()[:-len(topFolder)]
sys.path.append(moduleDirectory)

from reporting import *
%matplotlib inline
import matplotlib.pyplot as plt

path6 = r'C:\Users\ucqba01\Documents\Local Data\Round 6'

def plotRPS90(comparisondata,seriesList):
    sns.reset_defaults()
    
    rps90 = comparisondata.loc[comparisondata.waves==10000]
    rps90=rps90.loc[rps90.RPS==90]
    rps90.sort_values(by='fcost',inplace=True)
    rps90.set_index('fcost',inplace=True)

    sns.set_palette(sns.hls_palette(len(seriesList),l=0.4,s=0.9))
    rps90[seriesList].plot(marker='o',linestyle='-')
    plt.xlabel(r'Floating Cost %')
    plt.ylabel('Capacity (MW)')
    plt.gca().set_ylim(bottom=0)
    plt.legend()
    plt.title('Trends for RPS90WOFF')
    plt.show()
    plt.close()

    
round6 = makeComparisonTable(path6)
round6
```

```python
plotSeries=['Windoffshore_Floating','Windoffshore_Mid','Windoffshore_Shallow','Solar','Windonshore','NaturalgasOCGTnew','NaS']
plotRPS90(round6,plotSeries)
```
